```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('new csv/pcc_c2 cubic_red.csv')

# Drop rows with NaN values
data.dropna(inplace=True)

# Convert columns to numeric
data['Client 1_cubic'] = pd.to_numeric(data['Client 1_cubic'], errors='coerce')
data['Client _2_pcc'] = pd.to_numeric(data['Client _2_pcc'], errors='coerce')

# Drop rows with non-numeric values
data.dropna(inplace=True)

# Convert bits to megabytes per second (MBps)
data['Client 1_cubic'] /= 8 * 10**6
data['Client _2_pcc'] /= 8 * 10**6

# Plot the data
plt.plot(data['Interval start'], data['Client _2_pcc'], label='Client1_PCC', color='g')
plt.plot(data['Interval start'], data['Client 1_cubic'], label='Client2_CUBIC', color='r')

# Set x-axis range
plt.xlim(0, 100)

# Set x-axis ticks at intervals of 20 seconds up to 100 seconds
plt.xticks(range(0, 101, 20))

# Add title and labels
plt.title('')
plt.xlabel('Time in seconds')
plt.ylabel('Throughput (MBps)')

# Add legend inside the plot area
plt.legend(loc='upper right', fontsize='small')

# Display the plot
plt.savefig('/Users/shyamshrestha/Desktop/csv/new figure/final png/cubic-pcc')
plt.show()
```
